## Подключим необходимые библиотеки и определим глобальные константы

```python
import matplotlib.pyplot as plt
from typing import Dict, Iterable

ARG_N_FUNC = {
  0.0  : 1.000000,
  0.15 : 1.007568,
  0.3  : 1.031121,
  0.45 : 1.073456,
  0.6  : 1.140228,
  0.75 : 1.242129,
  0.9  : 1.400176,
  1.05 : 1.660300,
  1.2  : 2.143460
}

ARG = list(ARG_N_FUNC.keys())
FUNC = list(ARG_N_FUNC.values())
```

## Определим функцию для вычисления интеграла методом трапеций

```python
def trapezoidal(x: Iterable, f: Iterable) -> float:
  res, n = 0.0, len(x)

  for k in range(n - 1):
    res += (x[k + 1] - x[k]) * (f[k + 1] + f[k])

  return res / 2
```

## Вычислим интеграл методом трапеций

```python
plt.figure(figsize=(17, 10))
plt.plot(ARG, FUNC, marker='o', linestyle='dashed')
plt.grid()
plt.show()

I_h = trapezoidal(ARG, FUNC)
print(f'Интеграл, вычисленный методом трапеций с разбиением h: {I_h}')
```

## Вычислим интеграл методом трапеций с удвоенным разбиением

```python
plt.figure(figsize=(17, 10))
plt.plot(ARG[::2], FUNC[::2], marker='o', linestyle='dashed')
plt.grid()
plt.show()

I_2h = trapezoidal(ARG[::2], FUNC[::2])
print(f'Интеграл, вычисленный методом трапеций с разбиением 2h: {I_2h}')
```

## Вычислим интеграл методом трапеций с уточнением Ричардсона

```python
I_r = (4 * I_h - I_2h) / 3
print(f'Интеграл, вычисленный методом трапеций с уточнением Ричардсона: {I_r}')
```

## Определим функцию для вычисления интеграла методом Симпсона

```python
def simpson(x: Iterable, f: Iterable) -> float:
  res, n = 0.0, len(x)

  for k in range(n // 2):
    res += (x[2 * k + 1] - x[2 * k]) * (f[2 * k] + 4 * f[2 * k + 1] + f[2 * k + 2])

  return res / 3
```

## Вычислим интеграл методом Симпсона

```python
I_s = simpson(ARG, FUNC)
print(f'Интеграл, вычисленный методом Симпсона: {I_s}')
```
